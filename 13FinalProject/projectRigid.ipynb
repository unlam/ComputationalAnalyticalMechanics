## Joints

Joints are the connection points between the links of a robot. The most common ones, called low-order joints, have a single degree of freedom, either rotational or translational.

l01-03-l-joint-types-and-degrees-of-freedom-01.png

### Revolute joint

- **Revolute joints** operate on a single axis of rotation. These are driven by direct current motors that serve as actuators. In proportion to the electrical current that feeds them, they provide the required torque.

## Translational joints

- A **translational joint** adds a length to those of links 1 and 2 (those with lengths $\ell_1$ and $\ell_2$) to form the length $d$. The actuators for this type of joint are usually pneumatic or hydraulic in nature, where a piston performs the corresponding **linear force**.

Prismatic.png

- A **cylindrical joint** between links 1 and 2 that allows the rotation of 2 with respect to 1, and also the displacement of 2 along the axis of 1.  

CylindricalExample.png

For this simplified model we will assume that the joint is cylindrical, but only its translational degree of freedom will be used.

# Instructions
The arm will perform a sequence of movements in order to lift a piece from the ground and present it with the end effector at the tip of the arm facing upward in a determined position.

The stages are:
1. Arm vertical, retracted.
2. Rotate first section until pointing to element.
3. Extend until grasping element.
4. Rotate 1st section to form $60^\circ$ with the horizontal.
5. Rotate 2nd section until pointing upward.
6. Extend the 1st section completely until reaching the final position.

There are then six terminal stages between which there will be five distinct movements.

For each of the movements, the following is required:
- Determine the generalized coordinates and their change.
- Calculate the torques and/or forces of the motors as a function of time.
- Create a graphical representation of what was calculated.

## Considerations
We will make many simplifications to make the problem more accessible:
- Link 0: arm fixation to the ground
    - The first link of length $l_1$ rotates around point $O$ elevated $0.5\,\mathrm{m}$ from the ground.
- Links 2 and 3: solid homogeneous cylinders
    - aluminum material $\rho_{Al} = 2.7\,\mathrm{\frac{g}{cm^3}}$.
    - Lengths: equal $\ell_2 = \ell_3 = 2\,\mathrm{m}$.
    - diameters: equal $\emptyset_2 = \emptyset_3 = 10\,\mathrm{cm}$
    - Link 2 can fold inside link 1, which we will model as a hollow cylindrical tube.
        - Therefore $\ell_1 < d < \ell_1+ \ell_2$.
- Link 1: hollow cylindrical tube
    - length $\ell_1 = 2\,\mathrm{m}$
    - internal diameter $\emptyset_{1\,int} = \emptyset_2 = 10\,\mathrm{cm}$ so as to accommodate link 2 inside it
    - external diameter $\emptyset_{1\,ext} = 14\,\mathrm{cm}$
- Joints (and their motors)
    - We neglect their masses in the model, so they don't affect the dynamics except by providing torque or force.
    - Rotational speed around rotational joints $10\,\mathrm{\frac{°}{s}}$
    - Extension/retraction speed of link 2, $\dot{\ell}_2 = 5\,\mathrm{\frac{cm}{s}}$
- End effector
    - Position coincident with the end of the link of length $\ell_3$.
    - Mass concentrated at this point (point mass) $20\,\mathrm{kg}$.
- Piece to be lifted
    - We assume it to be point-like, that is, all its mass of $130\,\mathrm{kg}$ will be concentrated at the same point as the end effector position.
    - Initially located on the ground at $5.2\,\mathrm{m}$ from the center of the arm support bracket.
