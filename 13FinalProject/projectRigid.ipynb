## Uniones

Las uniones (joints) son los puntos de conexión entre los vínculos (links) de un robot.
Las más usuales, las llamadas de bajo orden, presentan un único grado de libertad, sea rotacional o traslacional.

![joint types](figuras/l01-03-l-joint-types-and-degrees-of-freedom-01.png)

### Unión rotacional

- Las **uniones rotacionales** (revolute joint) operan sobre un único eje de rotación. Estas son accionadas por motores de corriente continua que ofician de actuadores. En forma proporcional a la corriente eléctrica que se les alimenta provéen el **torque** requerido.

## Uniones traslacionales

- Una **unión traslacional** adiciona una longitud a las de los vínculos 1 y 2 (las de longitudes $\ell_1$ y $\ell_2$) para conformar la longitud $d$. Los actuadores sobre este tipo de uniones suelen ser de carácter neumático o hidráulico donde un pistón realiza la **fuerza lineal** correspondiente.

Prismatic.png

- Una **unión cilíndrica** entre los vínculos 1 y 2 que permite tanto la rotación de 2 respecto de 1, así como el desplazamiento de 2 a lo largo del eje de 1.  
![CylindricalExample](figuras/CylindricalExample.png)

Para este modelo simplificado asumiremos que la unión es cilíndrica, pero solo se hará uso de su grado de libertad traslacional.

# Enunciado
El brazo realizará una secuencia de movimientos con el objeto de levanta una pieza del suelo y presentarla con el efector en el extremo del brazo mirando hacia arriba en una determinada posición.
Los estadíos son
1. Brazo vertical, retraido.
1. Rotar sección primera hasta apuntar a elemento.
1. Extender hasta tomar elemento.
1. Rotar 1.a sección que forma $60^\circ$ con la horizontal.
1. Rotar 2.a sección hasta apuntar hacia arriba.
1. Extender la 1.a sección en su totalidad hasta alcanzar la posición final.

Hay entonces seis estadíos terminales entre los cuales habrá cinco distintos movimientos.

Se requiere para cada uno de los movimientos:
- Determinar las coordenadas generalizadas y su cambio.
- Calcular los torques y/o fuerzas de los motores en función del tiempo.
- Realizar una representación gráfica de lo calculado.

## Consideraciones
Haremos muchas simplificaciones para hacer más accesible el problema
- Vínculo 0: fijación del brazo al suelo
    - El primer vínculo de longitud $l_1$ rota en torno del punto $O$ elevado $0.5\,\mathrm{m}$ del suelo.
- Vínculos 2 y 3: cílindros homogéneos macizos
    - material aluminio $\rho_{Al} = 2.7\,\mathrm{\frac{g}{cm^3}}$.
    - Longitudes: iguales $\ell_2 = \ell_3 = 2\,\mathrm{m}$.
    - diámetros: iguales $\emptyset_2 = \emptyset_2 = 10\,\mathrm{cm}$
    - El vínculo 2 se podrá plegar dentro del vínculo 1, al que modelaremos como un tubo cílindrico hueco.
        - Entonces $\ell_1 < d < \ell_1+ \ell_2$.
- Vínculo 1: tubo cílindrico hueco 
    - longitud $\ell_1 = 2\,\mathrm{m}$
    - díametro interno $\emptyset_{1\,int} = \emptyset_2 = 10\,\mathrm{cm}$ de forma de poder alojar en su interior al vínculo 2
    - díametro externo $\emptyset_{1\,ext} = 14\,\mathrm{cm}$
- Uniones (y sus motores)
    - Despreciamos sus masas en el modelo, por lo que no afectan la dinámica salvo proveyendo torque o fuerza.
    - Velocidad de rotación en torno a uniones rotacionales $10\,\mathrm{\frac{^\circ}{s}}$
    - Velocidad de extensión/retracción del vínculo 2, $\dot{\ell}_2 = 5\,\mathrm{\frac{cm}{s}}$
- Efector
    - Posición coincidente con el extremo del vínculo de longitud $\ell_3$.
    - Masa concentrada en este punto (masa puntual) $20\,\mathrm{kg}$. 
- Pieza a levantar
    - La asumimos puntual, es decir toda su masa de $130\,\mathrm{kg}$ estará concentrada en un el mismo punto de la posición del efector.
    - Ubicado inicialmente en el suelo a $5.2\,\mathrm{m}$ del centro del soporte de la sujeción del brazo.
