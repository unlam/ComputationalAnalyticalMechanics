## Code to solve mechanics problems

- We'll write code using the general programming language [Python](https://www.python.org/)
- Writing code $\neq$ programming
    - **Programming**: Know how to analyze and mathematically model a problem (logic), devise and compare alternatives and know how to calculate which is the best considering context of use, constraints, architecture of the computer system, as well as interface issues with the human operator, etc.
        - It requires **a lot** of study and effort.
        - Every engineering student **can** learn to program.
    - **Writing code**: Translate a solution into a particular programming language.
        - It requires **little** study and effort.
        - Every engineering student **MUST** learn to write code.
- We will use a minimum of library functions
    - [SymPy](https://www.sympy.org/es/index.html) for algebra and symbolic analysis
    - [NumPy](https://numpy.org/) y [SciPy](https://scipy.org/) for numerical analysis
    - [matplotlib](https://matplotlib.org/) to generate visual representations of the results

## **Computational** analytical mechanics

- Those who take _ Computational analytical mechanics_...
    - ...they have already passed the __primary school__:
        - They showed that they understand and can do arithmetic calculations manually.

    - ...they have already passed algebra and calculus course at the __university__:
        - They demonstrated that they understand and can do integral and differential calculations manually as well as spatial (vector) and matrix calculations.

## JupyterLab: the environment for this course

- JupyterLab's _notebooks_ are displayed in a browser
- They are segmented into cells of different kinds
    - **markdown**: texts, figures, multimedia and symbols in standard mathematical notation.
    - **code**: symbolic calculation, graphing and numerical simulation.
- In these we'll distribute
    - theory
    - solved examples
    - exercises
- They are **reusable**, that is, you just need to copy and modify them
    - Thus an exercise is solved by slightly modifying an example
    - Those will be the means to perform and deliver exercises
    - Time wasted copying board $\rightarrow$ paper is avoided
    - Any errors are corrected and your re-run $\rightarrow$ code is never started from scratch

## [Flipped classroom:](https://en.wikipedia.org/wiki/Flipped_classroom) it changes the use of time

timeFlowchart.png

- Theory
    - students read, and try to apply, **before class**
    - doubts are clarified **in class**
    - if new questions arise, additional queries can be made **after the class**
- Exercises
    - they start **before class**
    - they are finished **in class**
    - corrections suggested by professors **after class**

| Time    | Traditional course | Flipped classroom |
| -------- | ------- |------- |
| Synchronous with professors<br />(in class)  | Emphasis on the exposition of new theory.<br />Only the remainder of time to attend to queries. |  Emphasis on finishing exercises for the week.<br /> When there is difficulty with one, the professor is available. |
| Asynchronous<br />(at home or wherever you can) | Part of it is a review of theory.<br />Most of it is doing the exercises.<br />When there is difficulty with one, there is no teacher to turn to. | Much of it is assimilating new theory.<br /> Begin the resolution of exercises.

## Organization of a flipped classroom

- This modality requires **a constant effort** on the part of the student
- If you wait for the synchronous meeting (class) to
  - read theory $\implies$ **time is not enough**
  - begin solving the exercises $\implies$ **not enough time**
  - catching up on the previous week's material $\implies$ **time is not enough**
- To help keep up to date there will be a **weekly evaluation**
  - **New worksheet** each week
  - **Mandatory submission** of two or more exercises in every synchronous encounter
- Of course, student can fail in one or more of the exercises.
  - There may have been a lack of understanding of some aspect of theory $\implies$ personal theory consultation and/or review
  - but such a flaw must be corrected in a **second submission**.

## Course material management: everything online

At UNLaM we employ the Microsoft Teams service as our CMS to track student's work, but any open source alternative, such as Moodle, can provide this service.  

- A team in [Microsoft Teams](https://teams.microsoft.com/) per semester
  - A channel for each **new topic each week** with:
   - **Theory**: in Jupyter notebooks
   - **Solved exercises**: notebook containing the code to solve them
   - **Worksheet**: pdf in the repository
   - **Homework**: two or more exercises from the worksheet to be submitted during class
   - **Asynchronous queries**
       - open so that others can benefit from the answers
       - 24/7 via the MS Teams App
       - they are answered during the week: **Don't wait for the synchronous meeting**
- A repository in [GitHub](https://github.com/unlam/ComputationalAnalyticalMechanics/) hosts the material sorted by topic/week
- All Jupyter notebooks are **reusable** and executable in [Google Colaboratory](https://colab.research.google.com/)

## Other cosiderations

The following aspects are specific to the course at UNLAM, and should be tuned considering your needs.

### Asynchronous online consultations: 24/7

| Activity    | Traditional course | Flipped classroom |
| -------- | ------- |------- |
| New theory is presented  | Synchronous | Asynchronous via Teams |
| Examples of exercise resolutions | Synchronous | Asynchronous via Teams |
| Queries about theory | Synchronous | Asynchronous via Teams and synchronous |
| Presentation of worksheet  | Synchronous | Asynchronous via Teams |
| Queries about exercises  | Synchronous | Asynchronous via Teams and synchronous |
| Exercises resolution<br />(by students)  | Asynchronous | Asynchronous (mainly)<br /> and synchronous (closing) |

### Conditions for a passing grade

| Activity    | Final grade share [%] | Observation |
| -------- | :-------: |------- |
| Exercises  | 60 | (see below) |
| Participate in forum | 20 | Answers that you share in Teams to queries from your classmates |
| Final project  | 20 | Resolution and exposition in front of teaching staff. |

To get a passing grade you MUST make the weekly deliveries of exercises.

### Exercises submission
- Each unit has a worksheet, out of which several will be requested for delivery, these will appear in _Teams_ under _Tasks_.

- The classes are for consultation and progress in the exercises, do not hesitate to ask us everything you require, so that you can make the most of us to evacuate your doubts.
    - It is very important that you take advantage of the classes for consultations and that you also write to us in Teams during the week asking for help if needed.
- On the day the unit is presented, you will be able to start looking at it and solving it, you will have the next class to make additional inquiries and you must deliver unfailingly before the next one begins.

### Delivery scheme
> The worksheet of class N is published in class N-1 $\rightarrow$ in class N you can seek help about it $\rightarrow$ at the end of class N all requested exercises must be submitted.

- After the day of delivery $\implies$ penalty on grading
    - That is why it is important that you try to advance on them, it is not necessary that they are perfect, once you send them we will correct them and we will provide feedback and ask you for the necessary modifications so that you can advance until you finish it satisfactorily.
    - When the course is over then you should have managed to do **ALL** the exercises that we marked as _Homework_.
