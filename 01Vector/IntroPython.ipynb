## Another type of variable: boolean

In addition to all the data we've seen, we also have a type of data called Boolean, or logical. They can take only two values: `True` or `False`.

```python
foo = True
print(foo)
print(not foo)
print(2*foo)
```

Comparisons produce Boolean variables

```python
print(2==1)
print(2==2)
print(6>6)
print(6>=6)
```

We can combine more than one condition

```python
print(2==3 or 3<9)
print(2==3 and 3<9)
print(foo or 6<1)
print(foo and 6<1)
```

## If conditional

`if` is a command that will allow us to tell _Python_, depending on some condition, to do or not to do something.

```python
a = 7
if a < 10:
  print('The number is less than 10.')
```

We can also tell it what to do if the condition is not met with the else command

```python
a = 7 
if a <= 5:
  print('the number is less or equal to 5')
else:
  print('the number is greater than 5')
```

We can combine for and if

```python
print(foo)
for i in foo:
  if i <= 5:
    print('the number ' + str(i) + ' is less or equal to 5')
  else:
    print('the number ' + str(i) + ' is greater than 5')
 
#Observe the indentation, it's super important.
```

Another example, now using the break command that breaks the for loop.

```python
print(foo)
for i in foo:
  if i <= 5:
    print(i)
  else:
    break
 
```

There are other loops that allow you to work in different ways than `for`. One of them is the `while` loop. The `while` loop is like the `for` loop but it will be repeated only as long as a condition is met.

```python
n = 0
while n <= 5 :
  print(n)
  n = n+1
```

It is important that you do not forget the line within the cycle that modifies the variable in the condition, otherwise you will stay forever inside the loop and we will have to go and rescue you. 

This can be useful to create the same list that we had created with range but in a different way. We can start with an empty list and take advantage of the append method that lists have, that adds something to the end of what we already had.

```python
thelist = []
n = 1
while n <= 5 :
  thelist.append(n)
  n = n+1 

print(thelist)
```

### Numpy
Fundamental library for anything related to numerical calculations, since it brings us the concept of vector or matrix that we did not have before and that we are going to use all the time.
We are also going to have at hand things like cosines, sines, numbers such as $\pi$ or $e$, and some functions such as solving eigenvectors and eigenvalues quickly that will be fundamental in this copurse (sorry for the spoiler).

To be able to use a library, we must first import it into our code and that is what will appear in the first line, then we use it.

Let's start with the widely used vectors or _arrays_.

```python
import numpy as np # We assign the short name np to Numpy. If we don't, every time we want to use cosine (for example), we'll have to type numpy.cos.

a = np.array([1,5,3,4,6]) # On the outside are the parentheses of the np.array function, on the inside are the brackets of my list. As we saw before, we read it from the inside out, we grab my list and turn it into an array.

print(a[0])
print(a[3])
print(a[-1])
print(a[1:4]) 
# Everything is the same as with lists so far

b = np.array([2,3,1,4,0])

print(a+b) # Sum of vectors (What if the lengths are different? Test it.)

print(a*b) # Product component by component, this is so because it will be more useful to have this at hand than the dot product, and ultimately if we want dot product we just have to add these components with sum().

prodesc = np.sum(a*b)
print(prodesc)

print(a+2) # We add 2 to each component.
```

Matrices, we build them using a list of lists.

```python
M = np.array([[1,2,3],[1,2,3],[1,2,3]])
print(M)
```

A different way to define vectors and arrays is with some Numpy functions.

```python
Z = np.zeros([2,5]) #Creates an array of zeros, 2 rows and 5 columns. What if I want a vector?
U = np.ones([2,5]) #Ditto but matrix of ones

print(Z)
print(U)
print(np.shape(Z))
print(M[:,0]) #the column
```

Try the command M.T  
What does it do?  
Be careful, _Python_ is case sensitive.

```python
L = M.T
print(np.dot(L,M)) #Matrix product
```

Other useful vectors that we're going to use a lot are those generated by Numpy's `linspace` and `arange` functions.
What is the difference between them?
Test them and you can tell.

```python
print(np.linspace(0, 1, 5)) # np.linspace(start, stop, number of components)
print(np.arange(0, 5, 0.5)) # np.arange(start stop, step)
```

### Matplotlib
A library for creating graphs, I think there is no need to explain the enormous usefulness of this.

Usually only the 'pyplot' module is _imported_ which is an API to the `matplotlib` functions.
This allows you to use a format very similar to that of the commercial software `MATLAB`.

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 4* np.pi, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z)
```

Let's write it with a few extra things, I leave you to see what each thing means.
Save your own copy and play with it.

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
z = np.cos(x)

plt.figure()
plt.plot(x, y, '-o', color='blue', label='sen(x)')
plt.plot(x, z, '-s', color='red', label='cos(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('trigonometric functions')
plt.grid()
```

There is another way to graph using Matlplotlib with identical results.
It is about creating the figure and the graph as if they were variables (they are called objects) and then changing the attributes of each object.
Both ways of graphing will be found in codes that we upload during the course. You pick the way you like the most to do things.

```python
fig,ax = plt.subplots(1) #Creates a figure and axes as variables.
ax.plot(x, y, '-o', color = 'b', label = 'sen(x)')
ax.plot(x, z, '-s', color='red', label='cos(x)')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('trigonometric functions')
ax.grid()
```
