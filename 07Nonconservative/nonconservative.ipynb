## Decomposition of generalized forces
For a system acted upon by external fields that produce potentials as functions of the coordinates, $V=V(q_i)$, the Euler-Lagrange equation can be obtained as

$$
\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q}} \mathcal{L} = \frac{\partial}{\partial q} \mathcal{L},
$$
where the derivative of the Lagrangian with respect to the coordinate is

$$
\frac{\partial}{\partial q} \mathcal{L} = - \frac{\partial}{\partial q} V = F^{c}.
$$
Since this is coming from a potential $V$, this is one of the **conservative forces**, $F^{c}$. 

But a good deal of forces exerted upon a system, such as friction or the force exerted by an engine, are **nonconservative forces**.
We know that $\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q}} \mathcal{L}$ is the equivalent of mass times acceleration, so, according to Newton's second law,

$$
m \ddot{\vec{r}} = \sum_{l=1}^L \vec{F}_l = \sum_{m=1}^M \vec{F}_m^{c} + \sum_{j=1}^J \vec{F}_j^{nc},
$$
is the sum of **all the $L$ applied forces**, the $M$ conservative, $\vec{F}^c$, and the $J$ nonconservative, $\vec{F}^{nc}$.

In the Euler-Lagrange equation corresponding to each coordinate $q_i$, we must add to the term related to the conservative forces, another term for the nonconservative forces.
Thus, a **generalized force** $Q_i$ related to forces that do not depend on a potential is added:
$$
\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q_i}} \mathcal{L} = \frac{\partial}{\partial q_i} \mathcal{L} + Q_i.
$$

### How to find the generalized force $Q_i$?
- The $Q_i$ for each $q_i$ are found from the _decomposition_ of a known force
- These don't need to be vectors. We have to estimate the _virtual work_ $\delta W$ that such force would do in a _virtual displacement_ along the coordinate $\delta q_i$.

### Review of work and energy

Through the displacement between positions 1 and 2, the sum of **nonconservative forces** exerted upon the system, $\vec{F^\text{nc}} = \sum_{j=1}^J \vec{F}_j$, can produce a change in the mechanical energy of the system, $E = T + V$.
This change is equal to the **work** done by these forces, $W_{1\rightarrow2}^\text{nc} = \Delta E$.

Work is not a function of the state of the system, that is, it does not depend on its configuration at 1 and at 2, instead, it depends on the trajectory from $1$ to $2$.
Thus, the path integral

$$
W_{1\rightarrow2}^\text{nc} = \int_1^2 \vec{F^\text{nc}} \cdot \mathrm{d}\vec{r} \neq \int_1^2 \mathrm{d} W_{1\rightarrow2}^\text{nc}
,
$$
can't be reduced to an integral of a hypothetical exact differential, $\mathrm{d} W_{1\rightarrow2}^\text{nc}$, since work is not a function of the position.
But if 1 and 2 are infinitesimally close to each other, there will be an infinitesimal work, $\delta W^\text{nc}$, that can be calculated through the possible **virtual displacements** $\delta \vec{r}$ from each point in space.

#### $\delta W^{nc}$ as a function of $Q_i$

Including the nonconservative external forces $\vec{F}^\text{nc}$ (the conservative forces are included through a potential $V$) we can calculate

$$
\delta W^\text{nc} = \vec{F}^\text{nc} \cdot \delta\vec{r}\,,
$$
where $\delta\vec{r}$ is a function of variations of the $n$ generalized coordinates $\delta q_i$, that is $\delta\vec{r} = \delta\vec{r} (\delta q_1, \delta q_2, \dots, \delta q_n)$.
This is called **work of virtual forces**, in this case, nonconservative forces.

Since each generalized coordinate $q_i$ is corresponded by a generalized force $Q_i$, a virtual displacement $\delta q_i$ of the coordinate produces the same variation of work:

$$
\delta W^\text{nc} = \sum_{i=1}^n Q_i \delta q_i
.
$$
Therefore, we can set

$$
\sum_{j=1}^J \vec{F_j} \cdot \delta\vec{r}_j = \sum_{i=1}^n Q_i \delta q_i,
$$
where $\delta\vec{r}_j$ is the **virtual displacement of the position vector where the nonconservative force, $\vec{F_j}$, is applied**.

### Example: pendulum with non-conservative external force

The string of a pendulum has certain elasticity, which must be considered when creating a model to obtain its dynamics.
This elasticity results in the length not being constant, but rather varying as a function of the forces acting on its end.
Therefore, a generalized coordinate, $x$, will be used to account for this stretching beyond the equilibrium length, $l_0$.
In the figure below, the weight of mass $m$ whose center is called $C$ is not only subject to the gravitational force $\vec{P} = m \vec{g}$ and the elastic force of the spring, but also experiences a force
$$
\vec{F}= F(t) \hat{I}
,
$$
which seeks to displace it from its equilibrium position.
This non-conservative force can be decomposed into two components, one tangential to the spring and another perpendicular to it.

### Generalized forces from virtual work

- Virtual displacement as a function of generalized coordinates

$$
\begin{aligned}
\delta \vec{r}_c
= \delta x \hat{\rho} &+ (l_0 + x) \delta \theta \hat{\theta} \\
= \delta x \left( \sin(\theta) \hat{I} - \cos(\theta) \hat{J} \right) &+ (l_0 + x) \delta \theta \left( \cos(\theta) \hat{I} + \sin(\theta) \hat{J} \right)
\end{aligned}
.
$$
- Virtual work due to $\vec{F}$ applied at $C$, and since it is directed toward $\hat{I}$, only the terms of $\delta \vec{r}_c$ in this direction will be remain when calculating this work

$$
\begin{aligned}
\delta W
&= \sum_{i=1}^n Q_i \delta q_i = \vec{F} \cdot \delta \vec{r}_C = F(t) \hat{I} \cdot \delta \vec{r}_c\\
&= F(t) \sin(\theta) \delta x + F(t) (l_0 + x) \cos(\theta) \delta \theta \\
&= Q_x \delta x + Q_\theta \delta \theta \\
&\implies 
\left\{
  \begin{array}{l}
    Q_x = F(t) \sin(\theta)\\
    Q_\theta = F(t) (l_0 + x) \cos(\theta)
  \end{array}\right.
\end{aligned}
$$

### Toward a more automatable procedure
> Reference  
> - $\S 6.3.2$ Transformation to generalized coordinates
Douglas Cline, Variational Principles in Classical Mechanics, 3rd ed. Rochester, NY 14627: University of Rochester River Campus Libraries, 2021  

The total variation of the work of such forces is
$$
\delta W^\text{nc} = \sum_{i=1}^n Q_i \delta q_i = \sum_{j=1}^J \vec{F}_j \cdot \delta \vec{ r}_j(\delta q_1,\dots,\delta q_n),
$$
where the forces $\vec{F}_j$ are applied at positions $\vec{r}_j$ which are vectors in some system (Cartesian, spherical, etc.), but which we express as functions of generalized coordinates, $q_i$.
Therefore, the variation of such positions $\delta \vec{r}_j$ is a function of the variations of the generalized coordinates $\delta q_i$.

An arbitrary virtual displacement $\delta \vec{r}_j$ can be related to a virtual displacement of the generalized coordinate $\delta q_i$ through the partial derivatives of the vector coordinates with respect to the generalized coordinates
$$
\delta \vec{r}_j = \sum_{i=1}^n \left( \frac{\partial \vec{r}_j}{\partial q_i} \right) \delta q_i
.
$$
It could be said that in $\vec{r}_j$ we explore how much such position would vary as a function of each $q_i$ through the derivative and then "advance" in that direction with $\delta q_i$, and in this way we carry out the virtual displacement, $\delta \vec{r}_j$.

Thus, the work we are looking for can be expressed as
$$
\delta W^\text{nc} = \sum_{i=1}^n Q_i \delta q_i = \sum_{j=1}^J \vec{F}_j \cdot \sum_{i=1}^n \left( \frac{\partial \vec{r}_j}{\partial q_i} \right) \delta q_i
.
$$

When determining the generalized force resulting from the non-conservative forces for one of the generalized coordinates, $Q_i$, the second summation is irrelevant, and we obtain
$$
Q_i \delta q_i = \sum_{j=1}^J \vec{F}_j \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right) \delta q_i,
$$
and it is possible to identify the generalized force $Q_i$ as
$$
\boxed{
Q_i = \sum_{j=1}^J \vec{F}_j \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right)
}.
$$

From here onward, the Euler-Lagrange equations considering all generalized forces can be written as

$$
\boxed{
\frac{\mathrm{d}}{\mathrm{d}t} \frac{\partial}{\partial \dot{q}_i} \mathcal{L}
- \frac{\partial}{\partial q_i} \mathcal{L}
=
\sum_k \lambda_k \frac{\partial}{\partial q_i} f_k
+ \sum_j \vec{F}_j \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right)
}
,
$$

which in cases where there are no constraint forces will be

$$
\frac{\mathrm{d}}{\mathrm{d}t} \frac{\partial}{\partial \dot{q}_i} \mathcal{L}
- \frac{\partial}{\partial q_i} \mathcal{L}
=
\sum_j \vec{F}_j \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right)
,
$$

### Generalized forces from the position where the force is applied

En este caso hay una única fuerza,
$\sum_{j=1}^J \vec{F}_j = \vec{F}(t)$,
y por tanto a las coordenadas generalizadas $x$ y $\theta$ corresponden sendas fuerzas generalizadas, $Q_x$ y $Q_\theta$. 

- The position vector of the point where $\vec{F}(t)$ is applied is written in the geometric coordinate system, but as a function of the generalized coordinates
$$
\vec{r}_C = (l_0 + x) \hat{\rho} = (l_0 + x) \left( \sin(\theta) \hat{I} - \cos(\theta) \hat{J} \right)
$$
- For each $q_i$, its $Q_i$ is calculated
$$
\begin{aligned}
Q_x
= \vec{F} \cdot \frac{\partial \vec{r}_C}{\partial x}
&= F(t) \hat{I} \cdot \left( \sin(\theta) \hat{I} - \cos(\theta) \hat{J} \right)
= F(t) \sin(\theta)\\
Q_\theta
= \vec{F} \cdot \frac{\partial \vec{r}_C}{\partial \theta}
&= F(t) \hat{I} \cdot \left[ (l_0 + x) \left( \cos(\theta) \hat{I} + \sin(\theta) \hat{J} \right) \right]
= F(t) (l_0 + x) \cos(\theta)
\end{aligned}
$$

```python
import sympy as sm # import functions for symbolic manipulations
from sympy.physics import mechanics as me
me.init_vprinting() # dot notation for velocity and acceleration
```

```python
e = me.ReferenceFrame('e', indices=['I', 'J', 'K'], latexs=['\mathbf{\hat{I}}', '\mathbf{\hat{J}}', '\mathbf{\hat{K}}'])
F = sm.Symbol('F')
F_vec = F*e.x
F_vec
```

```python
l0, zeta = sm.symbols('l_0, theta')
x = me.dynamicsymbols('x')
C_r = (l0 + x)* (sm.sin(zeta)* e.x + sm.cos(zeta)*e.y)
C_r
```

```python
x_Q = sm.Eq(
  sm.symbols('Q_x'),
  F_vec.dot(C_r.diff(x,e))
)
x_Q
```

```python
zeta_Q = sm.Eq(
  sm.symbols('Q_theta'),
  F_vec.dot(C_r.diff(zeta,e))
)
zeta_Q
```

## Another example: rod hanging from a cart
- Rod: $m_2$, $l$ (center of gravity at G), hanging from (A).
- Cart: $m_1$ linked to the wall through a spring of constant $k$.

erste.png

```python
# Physical parameters
m1, m2, k, g, l = sm.symbols('m_1, m_2, k, g, l', positive=True)
```

```python
# positions of the center of gravity of each object
e = me.ReferenceFrame('e')
x, zeta = me.dynamicsymbols('x, theta')
m1_r = x* e.x
m2_r = m1_r + (l/2)* (sm.sin(zeta)* e.x + sm.cos(zeta)* (-e.y) )
```

#### Kinetic energy

The mass of the rod is distributed, so it is not appropriate to calculate its kinetic energy as that of a particle, but rather as that of a rigid solid.
To calculate its corresponding rotational kinetic energy
$$
T_\mathrm{rotation} = \frac{1}{2} I \omega^2,
$$
we need the [moment of inertia](https://en.wikipedia.org/wiki/List_of_moments_of_inertia) $I$.
For a rod of length $l$ and mass $m$ rotating about an axis transverse to the longitudinal one that passes through its center of mass, it is
$$
I_\mathrm{rod\,cm} = \frac{m l^2}{12}.
$$
In this problem, such axis is displaced to one end of the rod, at $\frac{l}{2}$ from the center of mass, so the term indicated by the [Parallel-axis theorem](https://en.wikipedia.org/wiki/Parallel_axis_theorem) must be added
$$
I_\mathrm{rod\,end} = m \left(\frac{l}{2}\right)^2 + \frac{m l^2}{12} = m \frac{l^2}{3}.
$$

```python
rod_I_center = m2* l**2/ 12  # moment of inertia of the rod about an axis passing through its center
rod_I_end = rod_I_center + m2* (l/2)**2 # at (l/2) from the center, applying the parallel-axis theorem

rod_I =  sm.Eq(
  sm.Symbol('I_{rod}'),
  rod_I_end
)
rod_I
```

```python
oneHalf = sm.Rational(1, 2)
t = sm.symbols('t') # time
rod_T_rotational = sm.Eq(
  sm.Symbol('T_{rotational}'),
  oneHalf * rod_I.rhs* zeta.diff(t)**2
)
rod_T_rotational
```

On the other hand, for the cart there is no problem in considering the mass concentrated at the center of mass, since it is not subject to rotation. Therefore, we treat the kinetic energy of the cart in the usual way.

```python
def energyKineticTranslation(mass, position, frameOfReference):
    """
    Given a mass and the position of a point particle in a frame of reference, it returns its kinetic energy.
    
    Parameters
    ----------
    mass: (sympy.core.symbol.Symbol)
        of one particle
    position: (sympy.physics.vector.vector.Vector)
        of the particle's center of mass
    frameOfReference: (sympy.physics.vector.frame.ReferenceFrame)
        in which the position is expressed
    
    Returns
    -------
    Sympy Equality (sympy.core.relational.Equality)
        The kinetic energy of the system is made explicit on the right-hand side, as a function of the generalized coordinates and velocities, and time.
        Kinetic Energy, T = (m/2) \dot{\vec{r}} \cdot\dot{\vec{r}}
    """
    velocity = position.dt(frameOfReference)
    oneHalf = sm.Rational(1,2) # Rational: a quotient of integers
    T_translational = sm.Eq(
        sm.Symbol('T_\mathrm{translational}'),
        oneHalf* mass* velocity.dot(velocity)
        ).simplify()
    return T_translational
```

```python
cart_T = energyKineticTranslation(m1, m1_r, e)
rod_T_translational = energyKineticTranslation(m2, m2_r, e)
```

Therefore, the kinetic energy of the system is

```python
T = sm.Eq(
  sm.Symbol('T'),
  cart_T.rhs + rod_T_rotational.rhs + rod_T_translational.rhs )
T
```

#### Potential energy

The spring stores elastic potential energy.

```python
spring_V = sm.Eq(
  sm.Symbol('V_{elastic}'),
  oneHalf * k* x**2
)
spring_V
```

And although the rod is an extended body where its point of connection with the cart does not vary in height while the end does so in a pronounced way, we will make an approximation to simplify the problem. For the purposes of gravitational potential energy, it will be considered that the rod concentrates all its mass at its center of mass.

```python
def energyPotentialGravitational(mass, position, accelerationGravity):
    """
    Provides the gravitational potential energy of a particle of mass m and position r in a given frame of reference in which g is the acceleration due to gravity.
    
    Parameters
    ----------
    accelerationGravity: (sympy.physics.vector.vector.Vector)
        vector oriented according to the frame of reference
    mass: (sympy.core.symbol.Symbol
        of the body
    position: (sympy.core.symbol.Symbol)
        of the center of mass in a frame of reference related to the one for the acceleration due to gravity

    Returns
    -------
    Sympy Equality (sympy.core.relational.Equality)
        The potential energy of the system is made explicit on the right-hand side, as a function of the generalized coordinates and velocities, and time.
    coordinateGeneralized: Sympy Symbol (sympy.core.symbol.Symbol)
        V = - (m \vec{accelerationGravity}) \cdot{position}
    """
    V_gravitational = sm.Eq(
        sm.Symbol('V_\mathrm{gravitational}'),
        - (mass* accelerationGravity).dot(position)
        ).simplify()
    return V_gravitational
```

```python
accelerationGravity = g* (- e.y)
rod_V = energyPotentialGravitational(m2, m2_r, accelerationGravity)
rod_V
```

Then, the potential energy of the system is

```python
# Potential energy
V = sm.Eq(
  sm.Symbol('V'),
  spring_V.rhs + rod_V.rhs
)
V 
```

### Two generalized coordinates, two Euler-Lagrange equations

```python
def eulerLagrange(T, V, coordinateGeneralized):
    '''    
    Provides the Euler-Lagrange equation for one generalized coordinate starting from the energies.
    
    Parameters
    ----------
    T : Sympy Equality (sympy.core.relational.Equality)
        The kinetic energy of the system is explicit on the right-hand side, as a function of the generalized coordinates and velocities, and time.
    V : Igualdad Sympy (sympy.core.relational.Equality)
        The potential energy of the system is explicit on the right-hand side, as a function of the generalized coordinates and velocities, and time.
    coordinateGeneralized: Sympy Symbol (sympy.core.symbol.Symbol)
        for which we are looking the Euler-Lagrange equation
        
    Returns
    -------
    Sympy Equality (sympy.core.relational.Equality)
        Homogeneous Euler-Lagrange equation
    '''
    lagrangian = (T.rhs - V.rhs).expand()
    t = sm.Symbol('t') # since we have to differentiate respect to time, the function diff declares t as a symbol
    return sm.Eq(
        lagrangian.diff(coordinateGeneralized) 
        - lagrangian.diff(coordinateGeneralized.diff(t)).diff(t)
        , 0
    ).simplify()
```

From now on we will call these Euler-Lagrange equations the **homogeneous Euler-Lagrange equations** to differentiate them from the Euler-Lagrange equations that consider non-conservative forces.

$$
\frac{ \mathrm{d} }{ \mathrm{d} t } \frac{\partial \mathcal{L} }{ \partial \dot{x} } 
- \frac{\partial \mathcal{L} }{ \partial x } 
= 0
$$

```python
x_EL_homogeneous = eulerLagrange(T, V, x)
x_EL_homogeneous
```

$$
\frac{ \mathrm{d} }{ \mathrm{d} t } \frac{\partial \mathcal{L} }{ \partial \dot{\theta} } 
- \frac{\partial \mathcal{L} }{ \partial \theta } 
= 0
$$

```python
zeta_EL_homogeneous = eulerLagrange(T, V, zeta)
zeta_EL_homogeneous
```

### Now we add non-conservative forces
Two non-conservative forces are applied to the previous system:
- an external driving force $\vec{F}_\text{driving} = F(t) \hat{x}$
- a damping force proportional to velocity $\vec{F}_\text{damping} = - b \dot{x} \hat{x}$
both acting on the cart, so we consider that they are applied at the point position $\vec{r}_A$.

zweite.png

The variation of virtual work can be analyzed
$$
\begin{aligned}
\delta W^\text{nc}
& = \sum_j \vec{F}_j^\text{nc} \cdot \delta \vec{r}_j = \sum_i Q_i \delta q_i \\
& = \left[ \vec{F}_\text{damping} + \vec{F}_\text{driving} \right] \delta x + \left[ 0 \right] \delta \theta \\
& = \left[ - b \dot{x} + F(t) \right] \delta x + \left[ 0 \right] \delta \theta \\
& \implies Q_x = - b \dot{x} + F(t) \quad Q_\theta = 0
\end{aligned}
$$
But it may be simpler to use the expression
$$
Q_i = \sum_{j=1}^J \vec{F}_j \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right)
,
$$
which requires making explicit the position vector $\vec{r}_j$ where $\sum_{j=1}^J \vec{F}(t)$ is applied
$$
\vec{r}_A = x \hat{x}
$$
then
$$
\frac{\partial \vec{r}_A}{\partial x} = \hat{x} \qquad
\frac{\partial \vec{r}_A}{\partial \theta} = 0 ,
$$
and therefore
$$
Q_x
= \left[ \left(- b \dot{x} + F(t) \right) \hat{x} \right] \cdot \frac{\partial \vec{r}_A}{\partial x}
= \left[ \left(- b \dot{x} + F(t) \right) \hat{x} \right] \cdot \hat{x}
= - b \dot{x} + F(t)\\
Q_\theta
= \left[ \left(- b \dot{x} + F(t) \right) \hat{x} \right] \cdot \frac{\partial \vec{r}_A}{\partial \theta} = 0
$$
this last one is $0$ because $\vec{r}_A$ has no dependence on $\theta$.

```python
b = sm.Symbol('b', positive= True)
F = sm.Function('F')(t)
forces_m1_r = (-b* x.diff(t)+ F)* e.x # sum of all forces axerted upon (vector)
x_Q = forces_m1_r.dot(m1_r.diff(x, e))
zeta_Q = forces_m1_r.dot(m1_r.diff(zeta, e))
x_Q, zeta_Q 
```

The `eulerLagrange` function will be used to generate the homogeneous equation to be placed on the left side of the equality that will have the sum of generalized forces on the right side.
$$
\begin{aligned}
\frac{ \mathrm{d} }{ \mathrm{d} t } \frac{\partial \mathcal{L} }{ \partial \dot{x} }
- \frac{\partial \mathcal{L} }{ \partial x } 
&=  Q_x \\
\frac{ \mathrm{d} }{ \mathrm{d} t } \frac{\partial \mathcal{L} }{ \partial \dot{x} }
- \frac{\partial \mathcal{L} }{ \partial x } 
&=  -b \dot{x} + F(t) 
.
\end{aligned}
$$


```python
x_EL = sm.Eq(
  eulerLagrange(T, V, x).lhs,
  x_Q
)
x_EL
```

In the Euler-Lagrange equation for the coordinate $\theta$ there are no modifications

```python
zeta_EL = sm.Eq(
  eulerLagrange(T, V, zeta).lhs,
  zeta_Q
)
zeta_EL
```

In any case, it is clear that to obtain the dynamics one must solve this coupled system of these two differential equations.
In particular, one must take into account that $F = F(t)$, so this will have to be developed into an expression compatible with the resolution tools that are used. Toward the end of the course we will see the application of Fourier development for this type of problems.

## In summary: forces in the Lagrangian approach
If the force acting on the components of a system is
- caused by some **field** (electric, gravitational), it is considered by calculating the **potential** using $\vec{F}= - \vec{\nabla} V$,
- if it is of another origin, it is expressed as a **non-conservative** force, which can be dissipative forces, from an external motor, etc. They are written as generalized forces $Q_i$.
There are other forces, which are those that maintain a **constraint**.
For example, those exerted by a rigid bar maintaining a certain distance, or those exerted by a surface on which a component rests.
These do no work since the possible $\delta r$ of the system components will always be perpendicular to these (otherwise the constraint is not satisfied).
If we do not need their magnitude, they are not included in the calculation, but if we need it, this can be obtained with the technique of **Lagrange multipliers**.

### Euler-Lagrange con fuerzas de ligadura y no conservativas

$$
\boxed{
    \frac{\mathrm{d}}{\mathrm{d}t} \frac{\partial}{\partial \dot{q}_i} \mathcal{L}
    - \frac{\partial}{\partial q_i} \mathcal{L}
    =
    \sum_{k=1}^K \lambda_k (t) \frac{\partial}{\partial q_i} f_k
    + \sum_{j=1}^J \vec{F}_j (t) \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right)
}
\quad
\left\{
  \begin{array}{l}
  i= 1,2,\dots,n \quad \text{generalized coordinates} \\
  j= 1,2,\dots,J \quad \text{nonconservative forces}\\
  k= 1,2,\dots,K \quad \text{constraints}
  \end{array}
\right.
$$
where
$$
\sum_{k=1}^K \lambda_k(t) \frac{\partial}{\partial q_i} f_k,
$$
corresponds to the generalized forces coming from the $K$ constraint functions, $f_k$, and
$$
\sum_{j=1}^J \vec{F}_j (t) \cdot \left( \frac{\partial \vec{r}_j}{\partial q_i} \right),
$$
corresponds to the generalized forces produced by the $J$ nonconservative forces, $\vec{F}_j (t
)$.

This formulation allows us to use the `eulerLagrange` function that we have used up to this point to generate the homogeneous equation, that is, set equal to a `0` on the right side of the equality, `rhs`, but instead write the sum of the generalized forces resulting from the constraints and the non-conservative forces.
