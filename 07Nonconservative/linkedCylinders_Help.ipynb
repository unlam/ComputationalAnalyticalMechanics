### Consider the virtual works

With the given damping force, assuming $\hat{e}_y$ is vertical,

$$
\begin{aligned}
\delta W^\text{nc}
& = \sum_j \vec{F}_j^\text{ nc} \cdot \delta \vec{r}_j = \sum_i Q_i \delta q_i \\
& = - b \dot{\vec{r}} \cdot \delta \vec{r}^\text{ nc}\,, \\
\end{aligned}
$$
where $\delta \vec{r}^\text{ nc}$ is a virtual variation: of virtual works "where $\delta\vec{r}_j$ is the **virtual displacement of the coordinate where the external nonconservative force is applied**, $\vec{F}^{\text{ nc}}$"

The damping force depends on the velocity of the red dot shown in the following figure.

imagen.png

The angular velocity is the same for both cylinders, so:

$$
\vec{F}^\text{ nc} = -b (v_x \hat{e}_x + v_y \hat{e}_y) = - b \, v_y \hat{e}_y  = - b \, (R_2 \dot{\theta}) \hat{e}_y
$$

Now we consider the virtual displacement:

$$
\delta \vec{r}^{\text{nc}} = \frac{\partial \, \vec{r}^{\text{ nc}}}{\partial \theta} \delta \theta = R_2 \delta \theta
$$

Then we have

$$
\begin{aligned}
\delta W^{\text{ nc}} &= \vec{F}^\text{ nc} \cdot \left( \frac{\partial \, \vec{r}^{\text{ nc}}}{\partial \theta} \right) \delta \theta = Q_{\theta} \delta \theta\\
&= - b (R_2 \dot{\theta}) R_2 \delta \theta\\
&= - b \, R_2 ^2 \, \dot{\theta} \delta \theta
\end{aligned}
$$

Thus, $ Q_{\theta} = - b \, R_2 ^2 \, \dot{\theta} $

This is the generalized force that we have to add into the Euler-Lagrange equations:

$$
\frac{\mathrm{d}}{\mathrm{d}t} \frac{\partial}{\partial \dot{\theta}} \mathcal{L}
- \frac{\partial}{\partial \theta} \mathcal{L}
=
\sum_k \lambda_k \frac{\partial}{\partial \theta} f_k
+ Q_{\theta}\,.
$$

Here, we have the possibility of including constraints (Lagrange multipliers $\lambda_k$), if there are none, then that term vanishes. Do we have to include them in this problem?
