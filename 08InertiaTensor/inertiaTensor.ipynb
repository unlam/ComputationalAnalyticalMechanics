## Distribution and center of mass

- Each infinitesimal volume $d V$ has a $d m$
- Total mass 
$$
m = \int d m (\vec{r}) = \iiint \tilde{\rho}(\vec{r}) dV
$$

- If it's not homogeneous, the **density** $\tilde{\rho}(\vec{r})$ is a function of $\vec{r}$
- Center of mass (center of gravity)
$$
\vec{r}_{cm} = \frac{1}{m} \iiint \vec{r} \tilde{\rho}(\vec{r}) dV
$$


### The kinetic energy of each piece is a function of its velocity
$$
T = \int \frac{d m (\vec{r})}{2} \dot{\vec{r}}^2
$$
We are back with $\dot{\vec{r}}$ when there is a rotation.

#### Angular velocity of a spider over a frisbee

Viewing from above a frisbee rotating counterclockwise

aranaFrisbee.png

the angular velocity vector for the rotation $\dot{\varphi} = \frac{d \varphi}{d t}$ is directed upward. 

Viewing this in perspective with a position vector of the spider $\vec{r}$ on the frisbee that flies at a certain height starting from an origin fixed to the ground $O$.

lFig5.png

A small variation of the angle $\delta \varphi$ of the frisbee results in a respective small variation $\delta r$ of the spider's position with respect to that fixed point on the ground $\vec{r}_{a/o}$.
The algebraic calculation of the effect of $d \vec{\varphi} = d \varphi \hat{k}$ is the vector cross product of these factors, familiarly known as the *right-hand rule* for rotation
$$
d \vec{r}_{a/o} = d \vec{\varphi} \times \vec{r}_{a/f}
$$

The algebraic tool of the *right-hand rule* facilitates the analysis of $\vec{r}_{a/o}$ as a vector sum of the spider's position with respect to the frisbee's center of rotation $\vec{r}_{a/f}$ and the position of this with respect to the fixed point on the ground $\vec{r}_{f/o}$
$$
\begin{aligned}
\vec{r}_{a/o} &= \vec{r}_{f/o} + \vec{r}_{a/f} \\
d \vec{r}_{a/o} &= d \vec{r}_{f/o} + d \vec{\varphi} \times \vec{r}_{a/f} \\
\frac{d }{d t} \vec{r}_{a/o}
&= \frac{d }{d t} \vec{r}_{f/o} + \frac{d }{d t} \vec{\varphi} \times \vec{r}_{a/f} \\
\dot{\vec{r}}_{a/o}
&= \dot{\vec{r}}_{f/o} + \vec{\Omega}_{f/o} \times \vec{r}_{a/f}
\end{aligned}
$$
$$
\boxed{
\dot{ \vec{r} }_o = \frac{\mathrm{d}}{\mathrm{d} t} \vec{r}_o = \frac{\mathrm{d}}{\mathrm{d} t} \vec{R}_P + \vec{\Omega}_{P/o} \times \vec{r}_P = \dot{ \vec{R}}_P + \vec{\Omega}_{P/o} \times \vec{r}_P
}
$$
- $\vec{r}_o$ position with respect to the inertial system (ground)
- $\vec{R}_P$ position of the point where the rotation axis passes
- $\vec{r}_P$ point that rotates around $\vec{R}_P$ with $\vec{\Omega}_{P/o}$ such that $\vec{r}_o = \vec{R}_P + \vec{r}_P$

### Kinetic energy of a discrete system
Finite elements: divide into a discrete sum
$$
\begin{aligned}
T
&= \int \frac{d m (\vec{r})}{2} \dot{\vec{r}}^2
\simeq \sum_p \frac{m_p }{2} \dot{\vec{r}_p}^2
= \sum_p \frac{m_p }{2} \left( \dot{ \vec{R}} + \vec{\Omega} \times \vec{r}_p \right)^2 \\
&\simeq
\dot{ \vec{R}}^2 \sum_p \frac{m_p }{2}
+ \sum_p m_p \dot{ \vec{R}} \left( \vec{\Omega} \times \vec{r}_p \right)
+ \sum_p \frac{m_p }{2} \left( \vec{\Omega} \times \vec{r}_p \right)^2
\end{aligned}
$$
The middle term cancels if the origin is the center of mass
$$
\sum_p m_p \dot{ \vec{R}} \left( \vec{\Omega} \times \vec{r}_p \right) = \sum_p m_p \vec{r}_p \left( \dot{ \vec{R}} \times \vec{\Omega} \right) = \left( \dot{ \vec{R}} \times \vec{\Omega} \right) \sum_p m_p \vec{r}_p = 0
$$
The squared cross product can be rewritten (it involves many calculations)
$$
\left( \vec{\Omega} \times \vec{r}_p \right)^2
= \left( \vec{\Omega} \times \vec{r}_p \right) \cdot \left( \vec{\Omega} \times \vec{r}_p \right)
= \vec{\Omega}^2 \vec{r}_p^2 - \left( \vec{\Omega} \cdot \vec{r}_p \right)^2
$$
The kinetic energy can then be written as
$$
T \simeq \frac{m}{2} \left( \dot{\vec{R} } \right)^2 + \frac{1}{2} \sum_p m_p \left[ \vec{\Omega}^2 \vec{r}_p^2 - \left( \vec{\Omega} \cdot \vec{r}_p \right)^2 \right]
$$
where the first term is the translational (of the center of mass) and the second is the one corresponding to rotation.

### Rotational kinetic energy
We will use tensor notation as an alternative to vector notation.
- Vectors are summarized with a single subscript, understanding that a sum is performed over the three components
    - The position vector $\vec{r}_p$ is described by $r_i$
    - Similarly, $\vec{\Omega}$ is described by $\Omega_i$
- Since the sum over index $i$ is understood, the sum signs $\sum_i$ are not written
    - $\vec{A} \cdot \vec{B} = \sum_i A_i B_i$ is summarized as $A_i B_i$
    - $\vec{A}^2 = \vec{A} \cdot \vec{A}$ is summarized as $A_i A_i = A_i^2$
    - $\vec{A}^2 \vec{B}^2 = (\vec{A} \cdot \vec{A}) (\vec{B} \cdot \vec{B})$ is summarized as $A_i^2 B_k^2$ where different subscripts are used so it is not misinterpreted as $( \vec{A} \cdot \vec{B} )^2 = (A_i B_i)^2$

In the rotational kinetic energy, $\vec{\Omega}^2 \vec{r}_p^2$ is written as $\Omega_i^2 r_l^2$ using subscript $l$ in $r_l$ to differentiate it from $i$ in $\Omega_i$
$$
\begin{aligned}
T_\text{rot}
&\simeq \frac{1}{2} \sum_p m_p \left[ \vec{\Omega}^2 \vec{r}_p^2 - \left( \vec{\Omega} \cdot \vec{r}_p \right)^2 \right]
= \frac{1}{2} \sum_p m_p \left[ \Omega_i^2 r_l^2 - \Omega_i r_i \Omega_k r_k \right] \\
&\simeq \frac{1}{2} \sum_p m_p \left[ \Omega_i \Omega_k \delta_{ik} r_l^2 - \Omega_i \Omega_k r_i r_k \right]
= \frac{1}{2} \Omega_i \Omega_k \sum_p m_p \left[ r_l^2 \delta_{ik} - r_i r_k \right]
\end{aligned}
$$
where to be able to factor out the common term $\Omega_i \Omega_k$, use was made of the [Kronecker delta](https://en.wikipedia.org/wiki/Kronecker_delta) $\delta_{ii} =1, \delta_{ik} = 0\,(i \neq k)$ to make the substitution $\Omega_i^2 = \Omega_i \Omega_k \delta_{ik}$.

The **inertia tensor** is the quantity in brackets
$$
I_{ik} = \left[ r_l^2 \delta_{ik} - r_i r_k \right]
$$
where $r_l$ must be summed over the three vector components and $i$ and $k$ are the indices of the respective components.

### Inertia tensor
With the 2nd order tensor $I_{ik}$, we calculate $T = \frac{m}{2} \dot{\vec{R}}^2 + \frac{1}{2} I_{ik} \Omega_i \Omega_k$
$$
\overline{\overline{I}}
= I_{ik} =
\begin{pmatrix}
I_{xx} & I_{xy} & I_{xz} \\
I_{yx} & I_{yy} & I_{yz} \\
I_{zx} & I_{zy} & I_{zz} \\
\end{pmatrix}
=
\begin{pmatrix}
\sum_p m_p (y_p^2 + z_p^2) & - \sum_p m_p x_p y_p & - \sum_p m_p x_p z_p \\
- \sum_p m_p y_p x_p & \sum_p m_p (x_p^2 + z_p^2) & - \sum_p m_p y_p z_p \\
- \sum_p m_p z_p x_p & - \sum_p m_p z_p y_p & \sum_p m_p (x_p^2 + y_p^2) \\
\end{pmatrix}
$$
- The tensor $\overline{\overline{I}}$ is symmetric
- $I_{xx}, I_{yy}, I_{zz}$: **moments of inertia**, the others **products of inertia**

### Diagonalization of $I_{ik}$
- The $x_i$ that only leave the diagonal elements of $I_{ik}$ are **principal axes of inertia**, which are those of geometric symmetry of the system.
In this case
$T_\text{rot} = \frac{1}{2} \left( I_1 \Omega_1^2 + I_2 \Omega_2^2 + I_3 \Omega_3^2 \right)$
- If the rotation occurs about an axis of a system $x'_i$ that does not pass through one centered at the center of mass $x_i$, a tensor $I'_{ik}$ can be calculated from the one calculated from the center of mass $I_{ik}$
$$
I'_{ik} = I_{ik} + M [ a^2 \delta_{ik} - a_i a_k ]
$$
where $M$ is the mass of the entire system and $\vec{a}$ goes from the origin of this system to the center of mass.
This expression is that of the [****Steiner parallel axis theorem****](https://en.wikipedia.org/wiki/Parallel_axis_theorem).
Once this term is added, and if the axes of the coordinate system are parallel to the **principal axes of inertia**, there will be no products of inertia in $I_{ik}$.
For a vector that goes from the origin of the axis system $x'_i$ to the center of mass $\vec{a} = a_x \hat{x} + a_y \hat{y} + a_z \hat{z}$, we have that $a^2 = \vec{a} \cdot \vec{a} = a_x^2 + a_y^2 + a_z^2$.
Thus the Steiner term is
$$
M [ a^2 \delta_{ik} - a_i a_k ]
=
\begin{pmatrix}
M (a_y^2 + a_z^2) & - M a_x a_y & - M a_x a_z \\
- M a_y a_x & M (a_x^2 + a_z^2) & - M a_y a_z \\
- M a_z a_x & - M a_z a_y & M (a_x^2 + a_y^2),
\end{pmatrix}
$$
which is also a symmetric tensor.

### $I_{ik}$ Special cases
If the mass is distributed:
- linearly along $\hat{z} \implies x_p = y_p = 0$
$$
I_{xx} = I_{yy} = \sum_p m_pz_p^2 \quad I_{zz}= 0
$$
- in a plane $(x,y) \implies z_p = 0$
$$
I_{xx} = \sum_p m_p y_p^2 \quad I_{yy} = \sum_p m_p x_p^2\\
\quad I_{zz} = \sum_p m_p (x_p^2 + y_p^2) = I_{xx} + I_{yy}
$$

### Finally, the kinetic energy

Now we make the full expression for the kinetic energy explicit using vector notation. 
$$
\begin{aligned}
T
= \frac{m}{2} \dot{\vec{R}}^2 + \frac{1}{2} \overline{\overline{I}} \vec{\Omega}^2
= \frac{m}{2} \dot{\vec{R}}^{\mathrm{tr}} \dot{\vec{R}} &+ \frac{1}{2} \vec{\Omega}^{\mathrm{tr}} \overline{\overline{I}} \vec{\Omega}\\
T = \frac{m}{2} 
\begin{pmatrix}
\dot{R}_x & \dot{R}_y & \dot{R}_z \\
\end{pmatrix}
\begin{pmatrix}
\dot{R}_x \\
\dot{R}_y \\
\dot{R}_z \\
\end{pmatrix} 
&+
\frac{1}{2} 
\begin{pmatrix}
\Omega_x &
\Omega_y &
\Omega_z \\
\end{pmatrix} 
\begin{pmatrix}
I_{xx} & I_{xy} & I_{xz} \\
I_{yx} & I_{yy} & I_{yz} \\
I_{zx} & I_{zy} & I_{zz} \\
\end{pmatrix}
\begin{pmatrix}
\Omega_x \\
\Omega_y \\
\Omega_z \\
\end{pmatrix}
\\ 
T = \frac{m}{2} 
\left(
\dot{R}_x^2 + \dot{R}_y^2 + \dot{R}_z^2 
\right) 
&+
\frac{1}{2}
\begin{pmatrix}
\Omega_x &
\Omega_y &
\Omega_z \\
\end{pmatrix} 
\begin{pmatrix}
I_{xx} & I_{xy} & I_{xz} \\
I_{yx} & I_{yy} & I_{yz} \\
I_{zx} & I_{zy} & I_{zz} \\
\end{pmatrix}
\begin{pmatrix}
\Omega_x \\
\Omega_y \\
\Omega_z \\
\end{pmatrix} 
\\ 
\end{aligned}
$$

While this is the general form for the tensor $\overline{\overline{I}}$ with arbitrary orientation, usually the rotation that **diagonalizes** it is performed.
This can be done with the rotation matrix that is constructed with the eigenvectors of $\overline{\overline{I}}$ which indicates the axis rotation required so that in new ones, $\left( \hat{x}', \hat{y}', \hat{z}' \right)$, the solid presents geometric symmetry.
After that operation, the expression for kinetic energy is simpler

$$
\begin{aligned}
T = \frac{m}{2} \dot{\vec{R}}^{\mathrm{tr}} \dot{\vec{R}} &+ \frac{1}{2} \vec{\Omega'}^{\mathrm{tr}} \overline{\overline{I'}} \vec{\Omega'}\\
T = \frac{m}{2} 
\begin{pmatrix}
\dot{R}_x & \dot{R}_y & \dot{R}_z \\
\end{pmatrix}
\begin{pmatrix}
\dot{R}_x \\
\dot{R}_y \\
\dot{R}_z \\
\end{pmatrix}
&+
\frac{1}{2} 
\begin{pmatrix}
\Omega_x' &
\Omega_y' &
\Omega_z' \\
\end{pmatrix} 
\begin{pmatrix}
I_{x'x'} & 0 & 0 \\
0 & I_{y'y'} & 0 \\
0 & 0 & I_{z'z'} \\
\end{pmatrix}
\begin{pmatrix}
\Omega_x' \\
\Omega_y' \\
\Omega_z' \\
\end{pmatrix} 
\\
T = \frac{m}{2} 
\left(
\dot{R}_x^2 + \dot{R}_y^2 + \dot{R}_z^2 
\right) 
&+
\frac{1}{2}
\begin{pmatrix}
\Omega_x' &
\Omega_y' &
\Omega_z' \\
\end{pmatrix} 
\begin{pmatrix}
I_{x'x'} \Omega_x' \\
I_{y'y'} \Omega_y' \\
I_{z'z'} \Omega_z' \\
\end{pmatrix} 
\end{aligned}
$$

$$
\boxed{T = \frac{m}{2} 
\left(
\dot{R}_x^2 + \dot{R}_y^2 + \dot{R}_z^2 
\right) 
+
\frac{1}{2} 
\left(
I_{x'x'} \dot{\Omega}_x'^2 +
I_{y'y'} \dot{\Omega}_y'^2 +
I_{z'z'} \dot{\Omega}_z'^2 
\right)}
$$

since in tensor $\overline{\overline{I}}$, only the elements on the diagonal, the moments of inertia, are not null.
