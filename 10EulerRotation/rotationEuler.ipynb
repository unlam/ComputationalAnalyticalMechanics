## Euler angles
>**References**:
>- $\S$ 35 "Eulerian angles". Landau and Lifschitz. Mechanics. 3rd. ed. Elsevier  
>- Section 13.7 "The Eulerian Angles". Jerry B. Marion. Classical Dynamics of particles and systems. Academic Press.

LandauFig47.png

1. The system $x_i$ ($x_1, x_2, x_3$) is initially coincident with $X,Y,Z$.
1. Rotate an angle $\varphi$ applying $\dot{\varphi}$ on the $Z$ axis. This leaves $x_1$ along  $\overline{O N}$, the nodal line. 
1. Rotate an angle $\theta$ applying $\dot{\theta}$ on $\overline{O N}$. This rises $x_2$ from the plane $XY$ and makes $x_3$ to be no longer coincident with $Z$.
1. Rotate an angle $\psi$ applying $\dot{\psi}$ on $x_3$. This rises $x_1$ from the plane $XY$ and changes the direction of $x_2$.

Angles $\varphi, \theta, \psi$ are the **Euler angles** and will be the generalized coordinates used to describe any rotation.

The projection of $\dot{\varphi}, \dot{\theta}, \dot{\psi}$ over the angular velocities colinear with the $x_i$ are:

$$
\boxed{
\vec{\Omega} = 
\begin{pmatrix}
\Omega_1\\
\Omega_2\\
\Omega_3\\
\end{pmatrix}
= 
\begin{pmatrix}
\dot{\varphi} \sin{\theta} \sin{\psi} + \dot{\theta} \cos{\psi} \\
\dot{\varphi} \sin{\theta} \cos{\psi} - \dot{\theta} \sin{\psi} \\
\dot{\varphi} \cos{\theta} + \dot{\psi}
\end{pmatrix}
}
$$

## Euler's equations
>**References**:
>- $\S$ 36 "Euler's equations". Landau y Lifschitz. Mechanics. 3rd. ed. Elsevier.  
>- Section 13.8 "Euler's equations for a rigid body". Jerry B. Marion. Classical Dynamics of particles and systems. Academic Press.

The $x_i$ can be aligned with the body so that the inertia tensor $\overline{\overline{I}}$ is diagonal.
The elements of this diagonal are the **principal moments of inertia** of this body, $I_{i}$, and the $x_i$ are its **principal axes of inertia**.
In such case
$$
T_\text{rot} = \frac{1}{2} \left( I_1 \Omega_1^2 + I_2 \Omega_2^2 + I_3 \Omega_3^2 \right).
$$

We will solve the Euler-Lagrange equations for the generalized coordinates of rotation: $\varphi, \theta, \psi$.
For example, for the latter
$$
\frac{d}{d t} \frac{\partial T}{\partial \dot{\psi}}
- \frac{\partial T}{\partial \psi}
= 0
$$
To solve this we will use the projections in the $x_i$ system:
$$
\frac{d}{d t} \sum_i \frac{\partial T}{\partial \Omega_i} \frac{\partial \Omega_i}{\partial \dot{\psi}}
- \sum_i \frac{\partial T}{\partial \Omega_i} \frac{\partial \Omega_i}{\partial \psi}
= 0
$$
The $\frac{\partial \Omega_i}{\partial \psi}$ are
$$
\left\{
\begin{aligned}
\frac{\partial \Omega_1}{\partial \psi}
 &=
\dot{\varphi} \sin{\theta} \cos{\psi} - \dot{\theta} \sin{\psi} = \Omega_2\\
\frac{\partial \Omega_2}{\partial \psi}
 &=
 - \dot{\varphi} \sin{\theta} \sin{\psi} - \dot{\theta} \cos{\psi} = - \Omega_1\\
\frac{\partial \Omega_3}{\partial \psi}
 &= 0 \\
\end{aligned}
\right.
$$
while the $\frac{\partial \Omega_i}{\partial \dot{\psi}}$ are
$$
\left\{
\begin{aligned}
\frac{\partial \Omega_1}{\partial {\dot{\psi}} }
 &= \frac{\partial \Omega_2}{\partial {\dot{\psi}} }
 = 0 \\
\frac{\partial \Omega_3}{\partial {\dot{\psi}} }
 &= 1 \\
\end{aligned}
\right.
$$
And since $\frac{\partial T}{\partial \Omega_i} = I_i \Omega_i$ we get
$$
\frac{d}{d t} \left( I_3 \Omega_3 \right)
- \left[ I_1 \Omega_1 \Omega_2 + I_2 \Omega_2 (-\Omega_1) \right]
=
I_3 \dot{\Omega}_3
- (I_1 - I_2) \Omega_1 \Omega_2
 = 0
$$
Following the same procedure with $\theta$ and $\varphi$, the other two Euler-Lagrange equations are obtained.

Before summarizing all these equations, let us recall that the angular momentum $\vec{L} = \overline{\overline{I}} \vec{\Omega}$ can be decomposed into those associated with each $x_i$, that is $L_i$.
Each of these can be affected by some external torque $\vec{\tau}$ which must be decomposed into $\tau_i$ in each $x_i$.
We treat these the same as any other external generalized force $Q_i$:
$$
\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q_i}} \mathcal{L}
- \frac{\partial}{\partial q_i} \mathcal{L}
= Q_i
,
$$
for the case of rotation this can be summarized as
$$
\boxed{
I_i \dot{\Omega}_i
- (I_j - I_k) \Omega_j \Omega_k
= \tau_i
},
$$
which results in a system that must be solved together:
$$
\left\{
\begin{aligned}
 I_1 \dot{\Omega}_1
 - (I_2 - I_3) \Omega_2 \Omega_3
 &= \tau_1\\
 I_2 \dot{\Omega}_2
 - (I_3 - I_1) \Omega_3 \Omega_1
 &= \tau_2\\
 I_3 \dot{\Omega}_3
 -(I_1 - I_2) \Omega_1 \Omega_2
 &= \tau_3
\end{aligned}
\right.
$$

These equations will yield the same result as if the problem had been solved with the homogeneous Euler-Lagrange equations by adding the generalized force terms, which in this case are torques.
The choice of one method or another will depend on the complexity of the problem and the ease of identifying the generalized forces.
